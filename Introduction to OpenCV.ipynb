```python
#change the size of the image
img_ = cv2.resize(image,(300,300))
img_ = cv2.cvtColor(img_,cv2.COLOR_BGR2RGB)
plt.imshow(img_)
plt.show()
```

```python
#region of interest --- crop the image  ---slicing the array
crop=image[0:400,100:400]
img_ = cv2.cvtColor(crop,cv2.COLOR_BGR2RGB)
plt.imshow(img_)
plt.show()
```

```python
#draw diff shapes on images
white=np.zeros((512,512,3),np.uint8)+255  #pixel should be in int 
plt.imshow(white)
plt.show()
```

```python
image=cv2.imread(img)      
#real_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB) 
p1=(450,450)
p2=(10,10)
cv2.line(image,p1,p2,(255,0,0),20)   # to draw a line
#save image  
cv2.imwrite("line_selena.jpg",image)  #the notebook works in reverse format if working bgr format will save into rgb
plt.imshow(image)
plt.show()
```

```python
image=cv2.imread(img)
real_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
p1=(450,450)
p2=(10,10)
cv2.rectangle(image,p1,p2,(255,0,0),5)   # to draw a rectangle
cv2.imwrite("rect_sel.jpg",image)
plt.imshow(image)
plt.show()
```

```python
image=cv2.imread(img)
real_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
center=(250,250)
radius=250
cv2.circle(image,center,radius,(255,0,0),5)   # to draw a circle
cv2.imwrite("circle_sel.jpg",image)
plt.imshow(image)
plt.show()
```

```python
image=cv2.imread(img)
real_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
center=(250,250)
ax=(250,20)
a=30       # change the angle of the ellipse
start_=0   # this is the starting angle for the ellipse
end_=360   # this is the ending angle for the ellipse
cv2.ellipse(real_img,center,ax,a,start_,end_,(0,0,255),5)   # to draw a circle
plt.imshow(real_img)
plt.show()
```

```python
# write a text on image
image=cv2.imread(img)
real_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
text="Selena"
org=(0,500)
style=cv2.FONT_HERSHEY_SIMPLEX
size=5
cv2.putText(image,text,org,style,size,(255,0,0),5)   # writing text on image
cv2.imwrite("text_selena.jpg",image)
plt.imshow(image)
plt.show()
```
